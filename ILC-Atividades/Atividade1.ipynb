```python
#porcentagem de palavras que são verbos no infinitivo
def porcentagem(count, total):
    return (100*count)/total;

infinitivo = sorted(w for w in set(text1) if re.search('ar$', w) or re.search('er$', w) or re.search('ir$', w))

porcentagem(len(infinitivo), len(set(text1)))
```

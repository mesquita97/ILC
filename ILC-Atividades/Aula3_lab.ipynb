```python
import nltk.corpus
```

```python
from nltk.corpus import mac_morpho
```

```python
nltk.download('mac_morpho')
```

```python
mac_tagged = mac_morpho.tagged_words()
```

```python
tag_fd = nltk.FreqDist(tag for (word, tag) in mac_tagged)
```

```python
tag_fd.most_common()
```

```python
tag_fd.plot(cumulative='T')
%matplotlib inline
import matplotlib as plt
plt.rcParams['figure.figsize'] = (30,15)
```

```python
word_tag_pairs = nltk.bigrams(mac_tagged)
```

```python
noun_preceders = [a[1] for (a,b) in word_tag_pairs if b[1]=='N']
```

```python
fdist = nltk.FreqDist(noun_preceders)
```

```python
[tag for (tag,_) in fdist.most_common()]
```

```python
[tag for (tag,_) in fdist.most_common(1)]
```

```python
#verbos mais frequentes
cond_verb = nltk.ConditionalFreqDist((tag,word) for (word, tag) in mac_tagged)
list(cond_verb['V'])
```

```python
#substantivos mais frequentes
cond_noun = nltk.ConditionalFreqDist((tag,word) for(word,tag)in mac_tagged)
list(cond_verb['N'])
```

```python
#classes mais frequentes que a palavra melhor assume
cond_melhor = nltk.ConditionalFreqDist(mac_tagged)
cond_melhor['melhor'].most_common()
```
